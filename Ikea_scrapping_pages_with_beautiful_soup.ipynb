```python
#IMPORTING ALL LIBRARIES
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
#description
titles=[]
descriptions=[]
prices=[]
ratings=[]
```

```python
url="https://www.ikea.com/in/en/cat/tables-desks-fu004/?page=2"
```

```python
def items(container):
  items=container.find_all("div",class_="plp-fragment-wrapper")

  for item in items:
    title=item.find('span',class_="pip-header-section__title--small notranslate").text
    description=item.find('span',class_="pip-header-section__description-text").text
    price=item.find('span',class_="pip-temp-price__sr-text").text
    rating=item.find('span',class_="pip-rating__label").text

    #appending to the lists
    titles.append(title.strip())
    descriptions.append(description)
    prices.append(price)
    ratings.append(rating)

  #print(titles)

def get(url):
  try:
   response=requests.get(url)
   response.raise_for_status()
  except requests.exceptions.HTTPError as err:
   print(f"http error occured :{err}")
  except Exception as err:
   print(f'Other error occurred: {err}')
  soup=BeautifulSoup(response.text,"html.parser")
  #print(soup.prettify())
  container=soup.find("div",class_="plp-product-list__products")
  time.sleep(1)
  print(container.prettify())
  items(container)


```

```python
for i in range(1,8):
  url=f"https://www.ikea.com/in/en/cat/tables-desks-fu004/?page={i}"
  get(url)
```

```python
print(titles)
print(prices)
print(descriptions)
print(ratings)
print(len(titles))
print(len(prices))
print(len(descriptions))
print(len(ratings))
```

```python
#cleaning the data and converting into a uniform data type
for i in range(len(ratings)):
  prices[i]=prices[i].lstrip('Price Rs. ')
  prices[i]=int(prices[i])
  ratings[i]=ratings[i].lstrip('(')
  ratings[i]=ratings[i].rstrip(')')
  ratings[i]=int(ratings[i])


```
